```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
def searchARMA(data, max_p, max_q,minus = 0):
    max_p += 1
    max_q += 1
    aic = np.zeros((max_p,max_q))
    bic = np.zeros((max_p,max_q))
    for i in range(max_p):
        for j in range(max_q):
            model = ARIMA(data,order = (i,0,j)) 
            res = model.fit()
            aic[i,j] = res.aic -minus
            bic[i,j] = res.bic - minus
            print('p:',i,' q:', j,' aic:', aic[i,j], ' bic:', bic[i,j])
    return aic,bic
```

```python
for i in range(4):
    print(name_list[i],'\n')
    data = volume_list[i]
    aic_result,bic_result=searchARMA(data,3 ,3,minus=61100)
    print('min_aic: ',np.argmin(aic_result))
    print('min_bic: ',np.argmin(bic_result))
    print('\n')

    
```
